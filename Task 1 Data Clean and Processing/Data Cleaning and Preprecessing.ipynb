Outlier Treatment: -> Remove the outliers using boxplot

```python
def wisker(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
for i in ['Age','Fare']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

=====> Best Practice for Outlier Treatment

Outlier treatment is most appropriate for continuous variables (like Age and Fare), because extreme values can be due to errors, rare events, or data entry mistakes, and can distort analysis and models.

For discrete count variables (like SibSp and Parch), outlier treatment is less common. Their possible values are limited by real-world constraints (e.g., a person canâ€™t have 100 siblings on board), and high values might simply be rare but valid cases

=====> So, Should You Treat All Four Columns?

Definitely treat outliers in Age and Fare.

For SibSp and Parch:

If you see values that are clearly impossible (e.g., someone with 20 siblings on board), you can treat those as outliers.

If the high values are rare but possible, itâ€™s often better to leave them as-is, or just double-check for data entry mistakes.

Key Points

Why KNNImputer works here:
By encoding categories into integers and treating them as numerical features, KNN can find similar rows based on all features (including numerical ones like Age/Fare).

Rounding ensures valid categories:
After imputation, rounding guarantees the encoded values map back to valid original categories.

Handling high cardinality (e.g., Cabin):
If a column has too many unique values (like Cabin), group them (e.g., by deck: df['Cabin'] = df['Cabin'].str) before encoding.

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer


# Encode 'Cabin' and 'Embarked' into numerical values
categorical_cols = ['Cabin', 'Embarked']
encoders = {}

for col in categorical_cols:
    # Factorize to convert categories to codes (NaN becomes -1)
    codes, uniques = pd.factorize(df[col], use_na_sentinel=True)
    df[f'{col}_encoded'] = codes
    # Replace -1 (NaN) with np.nan for KNNImputer
    df[f'{col}_encoded'] = df[f'{col}_encoded'].replace(-1, np.nan)
    encoders[col] = uniques  # Save for inverse transform

# Select columns for imputation (include encoded categorical)
impute_cols = ['Cabin_encoded', 'Embarked_encoded']

# Initialize KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_imputed = pd.DataFrame(
    imputer.fit_transform(df[impute_cols]),
    columns=impute_cols
)

# Round encoded columns to integers and convert to nullable integer type
df_imputed[['Cabin_encoded', 'Embarked_encoded']] = (
    df_imputed[['Cabin_encoded', 'Embarked_encoded']]
    .round()
    .astype('Int64')  # Use pandas' nullable integer type
)

# Map encoded values back to categories
for col in categorical_cols:
    encoded_col = f'{col}_encoded'
    # Convert to integer and handle NaN safely
    df[f'{col}_imputed'] = df_imputed[encoded_col].apply(
        lambda x: encoders[col][int(x)] if pd.notna(x) else np.nan
    )

# Drop intermediate encoded columns
df = df.drop(columns=['Cabin_encoded', 'Embarked_encoded'])

# Display results
df[['Cabin', 'Cabin_imputed', 'Embarked', 'Embarked_imputed']]
```

```python
df.isnull().sum()
```

```python
# Replace original 'Cabin' and 'Embarked' with imputed values
df['Cabin'] = df['Cabin_imputed']
df['Embarked'] = df['Embarked_imputed']

# Drop the temporary imputed columns
df.drop(columns=['Cabin_imputed', 'Embarked_imputed'], inplace=True)

# Verify null counts
print(df.isnull().sum())
```
